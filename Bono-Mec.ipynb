```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fphi(w,phi0,t):
    return w*t + phi0
```

```python
def fr(r0,phi):
    exp = np.exp(phi)+np.exp(-phi)
    return (r0*exp)/2
```

```python
r0 = 0.0002
phi0 = 0
w = 0.12 #rad/s
t = np.linspace(0,120,10000)
phi = fphi(w,phi0,t)
r = fr(r0,phi)
```

```python
rx = r*np.cos(phi)
ry = r*np.sin(phi)
```

```python
plt.plot(rx, ry)
plt.grid()
plt.ylim(-200,200)
plt.xlim(-200,200)
plt.xlabel('m')
plt.ylabel('m')
plt.title(r'$r(\phi)$ con $\omega = 0.12$')
```
